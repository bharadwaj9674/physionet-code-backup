```python
import numpy as np
import matplotlib.pyplot as plt
import pywt
from scipy.io import loadmat

# Generate an example sine wave signal (replace this with your own signal)
fs = 100  # Sampling frequency (Hz)

file = loadmat('/Volumes/Bharadwaj/PhysioNet/i-care-international-cardiac-arrest-research-consortium-database-1.0/training/ICARE_0284/ICARE_0284_06.mat')
arr = file['val']
signal = arr[0]

# Generate time array
t = np.arange(len(signal)) / fs

# Perform Discrete Wavelet Transform (DWT)
wavelet = 'db4'  # Choose a wavelet function (Daubechies 4 in this case)
coeffs = pywt.wavedec(signal, wavelet)

# Plot the original signal and its DWT coefficients
plt.figure(figsize=(12, 8))

plt.subplot(len(coeffs) + 1, 1, 1)
plt.plot(t, signal)
plt.title('Original Signal')

for i, coeff in enumerate(coeffs):
    plt.subplot(len(coeffs) + 1, 1, i + 2)
    plt.plot(t[:len(coeff)], coeff)
    plt.title(f'Detail Coefficients {i + 1}')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pywt

# Generate example data (replace this with your actual signal data)
time = np.linspace(0, 10, 1000)
signal = np.sin(2 * np.pi * 5 * time) + 0.5 * np.random.randn(1000)  # Replace with your signal data

# Perform Discrete Wavelet Transform (DWT)
wavelet = 'db4'  # Choose a wavelet function
coeffs = pywt.wavedec(signal, wavelet)

# Plotting
plt.figure(figsize=(12, 8))
plt.subplot(len(coeffs) + 1, 1, 1)
plt.plot(time, signal)
plt.title('Original Signal')

for i, coeff in enumerate(coeffs):
    plt.subplot(len(coeffs) + 1, 1, i + 2)
    plt.plot(np.arange(len(coeff)), coeff)
    plt.title(f'Detail Coefficients {i + 1}')

plt.tight_layout()
plt.show()
```
