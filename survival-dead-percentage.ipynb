```python
import os
import re

list_good_poor = []

root_directory = '/Volumes/Bharadwaj/physionet-official/data/'
for folder_path, subfolders, files in os.walk(root_directory):
    for file_name in files:
        if file_name.endswith('.txt'):
            text_file_path = os.path.join(folder_path, file_name)
            
            with open(text_file_path, 'r') as text_file:
                content = text_file.read()
                print(content)
                pattern_outcome = r"Outcome: (\w+)"
                outcome = re.search(pattern_outcome, content).group(1)
                print(outcome)
                list_good_poor.append(outcome)
                
good_count = 0
poor_count = 0
for outcome in list_good_poor:
    if outcome == 'Good':
        good_count += 1
    if outcome == 'Poor':
        poor_count += 1
        
print(good_count)
print(poor_count)
```
