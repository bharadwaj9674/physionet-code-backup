```python
# limiting/ saving upto 5hz values and only positives included


import numpy as np
import matplotlib.pyplot as plt
from scipy.io import loadmat

# Load EEG data from the .mat file
file = loadmat('/Volumes/Bharadwaj/PhysioNet/i-care-international-cardiac-arrest-research-consortium-database-1.0/training/ICARE_0380/ICARE_0380_10.mat')
arr = file['val']

# Extract relevant EEG data from the first channel
first_channel_eeg = arr[0]  # Assuming the first channel is at index 0

# Perform Fourier Transform on the first channel
fft_result = np.fft.fft(first_channel_eeg)
freq_bins = np.fft.fftfreq(len(first_channel_eeg), 1 / 100)  # Frequency bins corresponding to FFT

# Find indices corresponding to the frequency range of 0-5 Hz
freq_indices = np.where((freq_bins >= 0) & (freq_bins <= 5))[0]

# Extract FFT values within the frequency range of 0-5 Hz
fft_values_0_to_5_hz = fft_result[freq_indices]

# Plot the magnitude of FFT results for the first channel
plt.figure(figsize=(8, 6))
plt.plot(freq_bins, np.abs(fft_result))
plt.xlabel('Frequency (Hz)')
plt.ylabel('Amplitude')
plt.title('Frequency Domain Representation (First Channel)')
plt.xlim(0, 50)  # Show frequencies up to 50 Hz for visualization
plt.grid()
plt.show()

# only 5hz data

# Print the extracted FFT values for 0-5 Hz
print("FFT values for 0-5 Hz:", fft_values_0_to_5_hz)

plt.plot(np.real(fft_values_0_to_5_hz[fft_values_0_to_5_hz>0]))
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.io import loadmat

file = loadmat('/Volumes/Bharadwaj/PhysioNet/i-care-international-cardiac-arrest-research-consortium-database-1.0/training/ICARE_0284/ICARE_0284_06.mat')
arr = file['val']

# Generate synthetic EEG data for 18 channels (replace with actual data loading)
num_channels = 18


sampling_rate = 100  # EEG sampling rate in Hz
duration = 300  # Duration of EEG recording in seconds
num_samples = sampling_rate * duration
time = np.arange(0, duration, 1/sampling_rate)

stack_all_channels = []

for i in range(num_channels):

    eeg_data = arr[i] #eeg data

    # Perform Fourier Transform for each channel
    fft_results = np.fft.fft(eeg_data)
    freq_bins = np.fft.fftfreq(num_samples, 1/sampling_rate)  # Frequency bins corresponding to FFT
    freq_indices = np.where((freq_bins >= 0) & (freq_bins <= 5))[0]
    fft_values_0_to_5_hz = fft_result[freq_indices]
    # plt.plot(fft_values_0_to_5_hz[fft_values_0_to_5_hz > 0]) -> plotting all signals above 5hz
    
    # stack_all_channels.append(fft_values_0_to_5_hz[fft_values_0_to_5_hz > 0])
    
    for i in fft_values_0_to_5_hz[fft_values_0_to_5_hz > 0]:
        stack_all_channels.append(i)

# stack_all_channels -> stacked frequencies of 1 patient, 18 channels from 0 to 5hz

print(len(stack_all_channels))
plt.figure(figsize=(30, 10))
plt.plot(stack_all_channels)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.io import loadmat

file = loadmat('/Volumes/Bharadwaj/PhysioNet/i-care-international-cardiac-arrest-research-consortium-database-1.0/training/ICARE_0284/ICARE_0284_06.mat')
arr = file['val']

# Generate synthetic EEG data for 18 channels (replace with actual data loading)
num_channels = 18


sampling_rate = 100  # EEG sampling rate in Hz
duration = 300  # Duration of EEG recording in seconds
num_samples = sampling_rate * duration
time = np.arange(0, duration, 1/sampling_rate)

stack_all_channels_noise_removed = []
stack_with_noise = []

for i in range(num_channels):

    eeg_data = arr[i] #eeg data

    # Perform Fourier Transform for each channel
    fft_results = np.fft.fft(eeg_data)
    freq_bins = np.fft.fftfreq(num_samples, 1/sampling_rate)  # Frequency bins corresponding to FFT
    freq_indices = np.where((freq_bins >= 0) & (freq_bins <= 5))[0]
    fft_values_0_to_5_hz = fft_results[freq_indices]
    # plt.plot(fft_values_0_to_5_hz[fft_values_0_to_5_hz > 0]) -> plotting all signals above 5hz
    
    # stack_all_channels.append(fft_values_0_to_5_hz[fft_values_0_to_5_hz > 0])
    
    fft_noise = []
    for i in fft_values_0_to_5_hz:
        if i > 0: 
            fft_noise.append(i)
            
    for i in fft_noise:
        stack_with_noise.append(i)
    
    
    fft_greater_point2hz = []
    for i in fft_values_0_to_5_hz:
        if i > 200000: #1e6 -> 0.2x1000000 , let say we considered just as noise
            fft_greater_point2hz.append(i)
    
    #stack_all_channels_noise_removed = []
    for i in fft_greater_point2hz:
        stack_all_channels_noise_removed.append(i)

# stack_all_channels -> stacked frequencies of 1 patient, 18 channels from 0 to 5hz

print(len(stack_all_channels_noise_removed))
plt.figure(figsize=(30, 10))
plt.plot(stack_all_channels_noise_removed)
plt.show()
mean_0284_noiseless = np.mean(stack_all_channels_noise_removed)
print(mean_0284_noiseless)
mean_0284_noise = np.mean(stack_with_noise)
print(mean_0284_noise)
plt.plot(stack_with_noise)
```
